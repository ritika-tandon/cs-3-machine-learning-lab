```
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
img=cv2.imread('opencv.png',1)
#img=cv2.line(img,(0,0),(255,255),(255,0,0),3)
img=cv2.line(img,(0,0),(255,255),(44,96,147),3)
#ARROW LINE
img=cv2.arrowedLine(img,(0,255),(255,255),(0,0,255),10)
#RECTANGE 
img=cv2.rectangle(img,(255,0),(280,128),(0,255,0),10)
#FILLED RECTANGLE
img=cv2.rectangle(img,(255,0),(280,128),(0,255,0),-1)
#CIRCLE
#img=cv2.circle(img,(100,200),63,(255,0,0),10)
img=cv2.circle(img,(100,200),63,(255,0,0),-1)
#PUT TEXT
font=cv2.FONT_HERSHEY_COMPLEX
img=cv2.putText(img,'open_cv',(40,50),font,2,(0,255,255),10)

cv2_imshow(img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
